```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros iniciales
d = 1
t = 3
a = d + t
d=100
b = 0
teta_o = 0
teta_i = teta_o
ni = 2
nt = 1

# Cálculo del ángulo crítico
teta_critico = np.arcsin(ni/ nt)
print(np.degrees(teta_critico))
print(teta_critico)
# Definición de la función de Snell
def snell(ni, nt, teta_i):
    """Calcula el ángulo de refracción usando la ley de Snell."""
    try:
        teta_t = np.arcsin(ni * np.sin(teta_i) / nt)
        return teta_t
    except ValueError:
        # Si el seno es mayor a 1, hay reflexión interna total
        return None  # Indica que no hay transmisión

# Llamada a la función Snell
teta_t = snell(ni, nt, teta_i)

# Salida
if teta_t is None:
    print("Ocurre reflexión interna total")
else:
    print(f"El ángulo de transmisión es {teta_t:.2f} rad")
```

```python
Rs=0
Rp=0
Ru=0
Tu=0
def fresnel(ni,nt,teta_t,teta_i):
     # Si hay reflexión interna total
    if teta_t is None:
        return 1
    Rs=((ni*np.cos(teta_i)-nt*np.cos(teta_t))/(ni*np.cos(teta_i)+nt*np.cos(teta_t)))**2
    Rp=((ni*np.cos(teta_t)-nt*np.cos(teta_i))/(ni*np.cos(teta_t)+nt*np.cos(teta_i)))**2
    Ru=(Rs+Rp)/2
    return (Ru)


x=fresnel(ni,nt,teta_t,teta_i)
print(x)
```

```python
# Inicialización de listas
b_values = np.linspace(0, d, 100)  # Valores de b desde 0 hasta d
Ru_values = []
Tu_values = []
teta_r_values = []  # Lista para teta_r

# Cálculo de Ru para cada b
for b in b_values:
    teta_i = np.arcsin(abs(b) / d)  # Cálculo de theta_i
    teta_t = snell(ni, nt, teta_i)  # Cálculo de theta_t con Snell
    Ru = fresnel(ni,nt,teta_t, teta_i)  # Cálculo de R_u con Fresnel
    Tu = 1 - Ru  # Cálculo de T_u
    teta_r=(ni*b)/(nt*d)

    Ru_values.append(Ru)
    Tu_values.append(Tu)
    teta_r_values.append(teta_r)

# Crear figura con dos subgráficos
fig, axs = plt.subplots(1, 2, figsize=(12, 5))  # 1 fila, 2 columnas

# Gráfica de Ru y Tu
axs[0].plot(b_values, Ru_values, label=r'$Ru$ (Reflexión)', color='b')
axs[0].plot(b_values, Tu_values, label=r'$Tu$ (Transmitancia)', color='r', linestyle='dashed')
axs[0].set_xlabel("b")
axs[0].set_ylabel("Valores")
axs[0].set_title("Reflexión y Transmitancia en función de $b$")
axs[0].legend()
axs[0].grid()

# Gráfica de teta_r
axs[1].plot(b_values, teta_r_values, label=r'$\theta_r$', color='g', linestyle='dotted')
axs[1].set_xlabel("b")
axs[1].set_ylabel(r"$\theta_r$")
axs[1].set_title(r"Ángulo $\theta_r$ en función de $b$")
axs[1].legend()
axs[1].grid()
```
